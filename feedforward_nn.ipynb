```python
from struct import unpack

import gzip
import numpy as np
import os

class FFNet:
    def __init__(self, dimensions):
        assert(len(dimensions) == 4)
        
        return
        
    # Temporary. Copied from external source.
    # https://github.com/hdmetor/NeuralNetwork/blob/master/data_load.py
    def read_images(self, file_name, data_folder):
        file_location = os.path.join(data_folder, file_name)

        with gzip.open(file_location, 'rb') as images:
            images.read(4)
            number_of_images = images.read(4)
            number_of_images = unpack('>I', number_of_images)[0]
            rows = images.read(4)
            rows = unpack('>I', rows)[0]
            cols = images.read(4)
            cols = unpack('>I', cols)[0]
            x = np.zeros((number_of_images, rows, cols), dtype=np.uint8) 

            print('[INFO]: Reading dataset ...')
            for i in range(number_of_images):
                for row in range(rows):
                    for col in range(cols):
                        tmp_pixel = images.read(1)  # Just a single byte
                        tmp_pixel = unpack('>B', tmp_pixel)[0]
                        x[i][row][col] = tmp_pixel
            print('[INFO]: Dataset loaded.')

        return x

    # Temporary. Copied from external source.
    # https://github.com/hdmetor/NeuralNetwork/blob/master/data_load.py
    def read_labels(self, file_name, data_folder):
        file_location = os.path.join(data_folder, file_name)

        with gzip.open(file_location, 'rb') as labels:
            labels.read(4)
            number_of_labels = labels.read(4)
            number_of_labels = unpack('>I', number_of_labels)[0]
            y = np.zeros((number_of_labels, 1), dtype=np.uint8) 

            for i in range(number_of_labels):
                tmp_label = labels.read(1)
                y[i] = unpack('>B', tmp_label)[0]

        return y

    # Load MNIST Data
    def load_mnist(self):
        X_test = read_images('t10k-images-idx3-ubyte.gz','MNIST_data')
        X_train = read_images('train-images-idx3-ubyte.gz','MNIST_data')
        y_test = read_labels('t10k-labels-idx1-ubyte.gz', 'MNIST_data')
        y_train = read_labels('train-labels-idx1-ubyte.gz', 'MNIST_data')
        print("[INFO]: X_test shape: ", X_test.shape)
        print("[INFO]: y_test shape: ", y_test.shape)
        print("[INFO]: X_train shape: ", X_train.shape)
        print("[INFO]: y_train shape: ", y_train.shape)

        return(X_test,X_train,y_test,y_train)


#X_test, X_train, y_test, y_train = load_mnist()
```
